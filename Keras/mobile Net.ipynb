```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

Mobile Net was originally trained on the imageNet library similary like VGG16

```python
mobile.summary()
```

Their are 2 ways in which a model is presented

SEQUENCIAL: This is a type of model in which we write the processes one after another and this is easy way but fails as it doesn't have the flexablity to go back to a layer that is mentioned above.

FUNCATONAL: This is a type of model in which is used for bigger and more complex models as it gives the freedom to go back to a previously used layer and is prefered over sequencial in MobileNets and Residual Neural Networks.

```python
x = mobile.layers[-6].output
output = Dense(units=29, activation='softmax')(x)
```

```python
model = Model(inputs=mobile.input, outputs = output)
```

```python
for layer in model.layers[:-23]:
    layer.trainable = False
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

The preprocessing proceeds as so that first we get the location of the files. In line 2 we convert the image to array and then 3rd line is to expand the dimesions of a image to the size neede by the mobilenet image. Then we finally pass this image to the preprocessing of the mobilenet which is that the values between 0 and 255 are converted to -1 to 1. 

```python
def preprocess(file):
    loc = r'C:/Users/suyash/Desktop\asl_alphabet_test\asl_alphabet_test'
    img = image.load_img(loc+'\\'+file, target_size=(224,224))
    img_array = image.img_to_array(img)
    img_dims = np.expand_dims(img_array, axis=0)
    return tf.keras.applications.mobilenet.preprocess_input(img_dims)
```

All the file and folder organisation is done by the fileconvert code take a look int o that too

```python
trainData = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input)\
    .flow_from_directory(directory=TRAIN,target_size=(224,224), batch_size=3)
    
testData = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input)\
    .flow_from_directory(directory=TEST, target_size=(224,224), batch_size=28)
```

```python
model.fit(x=trainData, validation_data=0.33, epochs=1, verbose=1)
```
